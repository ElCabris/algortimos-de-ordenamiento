```python
import matplotlib.pyplot as plt
from random import randint
from time import time
from Bubble_Sort import bubble_sort
from Insertion_Sort import insertion_sort
from QuickSort import quick_sort
from Selection_Sort import selection_sort
```

```python
a = []
x = [] # cantidad de elementos en la lista
y = [[], [], [], []] # tiempo de total requerido para ordenar la lista
funciones = [bubble_sort, insertion_sort, quick_sort, selection_sort]

for i in range(0, 10000, 100):
    x.append(i)
    a.clear()
    for _ in range(i):
        a.append(randint(0, i))


    for j in funciones:
        lista_entrada = a[:]
        t_inicial = time()
        if j == quick_sort:
            j(lista_entrada, 0, len(a) - 1)
        else:
            j(lista_entrada)
        t_final = time()
        t_total = t_final - t_inicial

        if j == bubble_sort:
            y[0].append(t_total)
        elif j == insertion_sort:
            y[1].append(t_total)
        elif j == quick_sort:
            y[2].append(t_total)
        elif j == selection_sort:
            y[3].append(t_total)
```

```python
plt.plot(x, y[0])
plt.xlabel("Elementos en la lista")
plt.ylabel("Tiempo de ejecución")
plt.title("Bubble Sort")
plt.show()
```

```python
plt.plot(x, y[1])
plt.xlabel("Elementos en la lista")
plt.ylabel("Tiempo de ejecución")
plt.title("Insertion Sort")
plt.show()
```

```python
plt.plot(x, y[2])
plt.xlabel("Elementos en la lista")
plt.ylabel("Tiempo de ejecución")
plt.title("Quick Sort")
```

```python
plt.plot(x, y[3])
plt.title("Selection Sort")
plt.xlabel("Elementos en la lista")
plt.ylabel("Tiempo de ejecución")
plt.show()
```

```python
plt.plot(x, y[0], label="Bubble Sort")
plt.plot(x, y[1], label="Insertion Sort")
plt.plot(x, y[2], label="Quick Sort")
plt.plot(x, y[3], label="Selection Sort")
plt.xlabel("Elementos en la lista")
plt.ylabel("Tiempo de Ejecución (Segundos)")
plt.legend()
plt.show()
```
